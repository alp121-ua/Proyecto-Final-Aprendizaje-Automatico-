# Tratamiento de Columnas Categóricas

Este notebook aplica **one-hot encoding** a las variables categóricas del dataset, transformándolas en columnas binarias aptas para modelos de machine learning.

## 1. Importación de librerías

Se importa `pandas` para la manipulación del dataset.

```python
import pandas as pd
```

## 2. Carga del dataset

Se carga el CSV de entrada, que contiene el dataset sin la variable target y con las columnas categóricas aún sin codificar.

```python
# Cargar datos
df = pd.read_csv('base-sin-target.csv')
```

## 3. Definición de la función `one_hot`

Función auxiliar que aplica `pd.get_dummies` sobre una columna concreta:
- **`drop_first=True`**: elimina la primera categoría para evitar multicolinealidad perfecta.
- **`dtype=int`**: las columnas resultantes se codifican como enteros (0/1) en lugar de booleanos.

```python
# One-hot encoding
def one_hot(df, columna):
    return pd.get_dummies(df, columns=[columna], drop_first=True, dtype=int)
```

## 4. Aplicación del encoding

Se aplica one-hot encoding sobre las cinco variables categóricas del dataset:
- `payment_type`: tipo de pago utilizado.
- `source`: canal de origen de la solicitud.
- `device_os`: sistema operativo del dispositivo.
- `employment_status`: situación laboral del solicitante.
- `housing_status`: situación de vivienda del solicitante.

```python
df = one_hot(df, 'payment_type')
df = one_hot(df, 'source')
df = one_hot(df, 'device_os')
df = one_hot(df, 'employment_status')
df = one_hot(df, 'housing_status')
```

## 5. Exportación del dataset tratado

Se guarda el dataset con las variables categóricas ya codificadas como `base-categoricas-tratado.csv`, listo para el siguiente paso del pipeline (tratamiento de outliers).

```python
df.to_csv("base-categoricas-tratado.csv", index=False)
```
